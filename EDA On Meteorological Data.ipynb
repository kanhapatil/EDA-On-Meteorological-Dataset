***getting started with the model***

***importing required libraries/packages***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### <h3 style="background-color:black;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Importing & Reading the dataset 📝 </centre></strong></h3>

```python
df_weather = pd.read_csv("weatherHistory.csv")
```

```python
df_weather_row_count, df_weather_column_count=df_weather.shape
print('Total number of rows:', df_weather_row_count)
print('Total number of columns:', df_weather_column_count)
```

### <h3 style="background-color:black; color:white; text-align: center; padding-top: 5px; padding-bottom: 5px; border-radius: 15px 50px;"><strong><centre>Checking for datatypes, unique values & missing values✏️ </centre></strong></h3>

```python
df_weather.info()
```

```python
print ("Unique values are:\n",df_weather.nunique())
```

```python
Summary_Weather = df_weather["Summary"].value_counts().reset_index()
Summary_Weather.columns = ["Weather Type", "count"]
Summary_Weather
```

```python
wt_missing = df_weather.isna().sum()
wt_missing 
```

***Function to find the percentage of missing values***

```python
def missing_percentage(col):
    return df_weather[col].isna().sum() / df_weather.shape[0] * 100

missing_percentage("Precip Type")
```

```python
df_weather['Precip Type'].fillna(df_weather['Precip Type'].value_counts().index[0],inplace=True)
df_weather.isna().sum()
```

```python
df_weather.isna().sum()
```

```python
df_weather.head()
```

***checking Datatypes***

```python
df_weather.dtypes
```

***Importing some another libraries (datetime & timedelta)***

```python
import datetime as dt
from datetime import timedelta
```

```python
df_weather["Formatted Date"] = pd.to_datetime(df_weather["Formatted Date"])
```

```python
plt.figure(figsize=(12,7))
plt.xticks(rotation=90)
sns.barplot(data=df_weather, x="Summary", y="Temperature (C)",hue="Precip Type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.xticks(rotation=90)
sns.lineplot(data=df_weather, x="Summary", y="Pressure (millibars)", hue="Precip Type", style="Precip Type")
plt.show()
```

```python
sns.pairplot(df_weather,hue="Humidity")
plt.show()
```
